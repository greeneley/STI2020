```python
class congruences_linaire:
    def __init__(self, x_0, a, c, m, N):
        self.x_0 = x_0
        self.a = a
        self.c = c
        self.m = m
        self.N = N 
        
    def entier_aleatoire(self):
         X = []
         X.append(self.x_0)
         for n in range(1,self.N):
            X.append((self.a*X[n-1] + self.c) % self.m)
         return X
    def valeur_aleatoire(self):
         X = []
         R = []
         X.append(self.x_0)
         for n in range(1,self.N):
            X.append((self.a*X[n-1] + self.c) % self.m)
            R.append( X[n] / self.m )
         return R
```

```python
cas_1 = congruences_linaire(3, 5, 3, 16, 16)
print(cas_1.entier_aleatoire())
print(cas_1.valeur_aleatoire())
```

```python
cas_2 = congruences_linaire(1, 3, 0, 7, 100)
print(cas_2.entier_aleatoire())
print(cas_2.valeur_aleatoire())
```
